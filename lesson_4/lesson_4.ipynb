### 1. 
Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию.

### Решение.

```python
a = 200
b = 800

print('Среднее:', (a + b) / 2)
print('Дисперсия:', (b - a)**2 / 12)
```

### 2. 
О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
Если да, найдите ее.

### Решение.

```python
Dx = 0.2
a = 0.5
#Из формулы дисперсии Dx = (b-a)**2/12 выводим формулу правой границы: b = (Dx*12)**0.5 + a
b = (Dx * 12)**0.5 + a
Mx = (a + b) / 2

print('Правая граница:', b)
print('Среднее:', Mx)
```

### 4. 
Рост взрослого населения города X имеет нормальное распределение.<br>
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.<br>
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:<br>
а). больше 182 см<br>
б). больше 190 см<br>
в). от 166 см до 190 см<br>
г). от 166 см до 182 см<br>
д). от 158 см до 190 см<br>
е). не выше 150 см или не ниже 190 см<br>
ё). не выше 150 см или не ниже 198 см<br>
ж). ниже 166 см.

### Решение.

Используя формулу $Z = \left(\frac{X - \mu}{\sigma}\right)$ и таблицу накопленного нормального распределения получаем:

```python
def z(x):
    Mx = 174
    sigma = 8
    return (x - Mx) / sigma

print('Z(150) =', z(150))
print('Z(158) =', z(158))
print('Z(166) =', z(166))
print('Z(182) =', z(182))
print('Z(190) =', z(190))
print('Z(198) =', z(198))
```

```python
# Z(182) = 1.0
print('а). больше 182 см:', 1 - 0.8413)

# Z(190) = 2.0
print('б). больше 190 см:', 1 - 0.9772)

# Z(166) = -1.0
# Z(190) = 2.0
print('в). от 166 см до 190 см:', 0.9772 - 0.1587)

# Z(166) = -1.0
# Z(182) = 1.0
print('г). от 166 см до 182 см:', 0.8413 - 0.1587)

# Z(158) = -2.0
# Z(190) = 2.0
print('д). от 158 см до 190 см:', 0.9772 - 0.0228)

# Z(150) = -3.0
# Z(190) = 2.0
print('е). не выше 150 см или не ниже 190 см:', 0.0014 + 1 - 0.9772)

# Z(150) = -3.0
# Z(198) = 3.0
print('ё). не выше 150 см или не ниже 198 см:', 0.0014 + 1 - 0.9986)

# Z(166) = -1.0
print('ж). ниже 166 см.:', 0.1587)
```

### 5. 
На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от
математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

### Решение.

```python
x = 190
Mx = 178
Dx = 25
sigma = Dx**0.5

print('Отклонение в сигмах:', (x - Mx)/sigma)
```
